```python
#exercise set2 
#compare elements of 2 panda series 
import pandas as pd

series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

print("Equal:", series1 == series2)
print("Greater than:", series1 > series2)
print("Less than:", series1 < series2)
```

```python
#2.add, subtract, multiply, divide two series
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

print("Addition:", series1 + series2)
print("Subtraction:", series1 - series2)
print("Multiplication:", series1 * series2)
print("Division:", series1 / series2)
```

```python
#3.convert dictionary to panda series
dict1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
s = pd.Series(dict1)
print(s)
```

```python
#4.convert panda series to array
s = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
array = s.to_numpy()
print(array)
```

```python
#display most frequent value in series
import pandas as pd

df = pd.read_csv('AfricaCupOfNationsMatches.csv')
most_frequent = df['HomeTeamGoals'].mode()[0]
df['HomeTeamGoalsModified'] = df['HomeTeamGoals'].apply(lambda x: x if x == most_frequent else 'Other')
print(df[['HomeTeamGoals', 'HomeTeamGoalsModified']])
```
